````python
# New function to compare components using GPT-4
def compare_components_gpt4(components1, components2, package_name):
    """
    Compares components from two PL/SQL packages using GPT-4 and generates detailed comparison reports.

    Parameters:
    - components1: Dictionary of components from the first package (e.g., HERITAGE)
    - components2: Dictionary of components from the second package (e.g., NEW_GEMINIA)
    - package_name: Name of the package being compared

    Returns:
    - differences: A dictionary containing added, removed, and modified components for each component type.
    """
    logging.info("Comparing components using GPT-4.")
    differences = {}
    reports_output_dir = os.path.join('gpt_reports', package_name)
    os.makedirs(reports_output_dir, exist_ok=True)

    for comp_type in components1.keys():
        set1 = set(components1[comp_type].keys())
        set2 = set(components2[comp_type].keys())

        added = set2 - set1
        removed = set1 - set2
        modified = set()

        for common in set1 & set2:
            content1 = components1[comp_type][common].strip()
            content2 = components2[comp_type][common].strip()
            if content1 != content2:
                modified.add(common)
                # Generate comparison report using GPT-4
                report = compare_plsql_procedures_gpt4(common, content1, content2, comp_type, package_name)
                if report:
                    # Save the report to a markdown file
                    report_file_path = os.path.join(reports_output_dir, f"{comp_type}_{common}_comparison_report.md")
                    with open(report_file_path, 'w', encoding='utf-8') as f:
                        f.write(report)
                    logging.info(f"Comparison report for {comp_type} '{common}' saved to '{report_file_path}'.")
                else:
                    logging.error(f"Failed to generate comparison report for {comp_type} '{common}'.")

        differences[comp_type] = {
            'added': list(added),
            'removed': list(removed),
            'modified': list(modified)
        }
    logging.info("Finished comparing components using GPT-4.")
    return differences

def compare_plsql_procedures_gpt4(proc_name, proc_code1, proc_code2, comp_type, package_name):
    """
    Compares two PL/SQL components (procedures, functions, etc.) using GPT-4 and returns a formatted report.

    Parameters:
    - proc_name: Name of the component
    - proc_code1: Code of the component from the first package
    - proc_code2: Code of the component from the second package
    - comp_type: Type of the component (e.g., 'procedures', 'functions')
    - package_name: Name of the package being compared

    Returns:
    - report: A string containing the formatted comparison report
    """
    # Construct the prompt
    prompt = f"""
You are an expert PL/SQL developer tasked with comparing two versions of a PL/SQL procedure 
and generating a detailed report highlighting the differences. 
Your primary goal is to ensure accuracy in identifying changes, 
especially in business logic, to prevent regression issues and facilitate easy merging of the two versions.

Here are the details of the two procedure versions you need to compare:

Procedure Name: <proc_name>{{proc_name}}</proc_name>
Package Name: <package_name>{{package_name}}</package_name>

Version 1 ({package_name}_HERITAGE):
```sql
{proc_code1}
```

Version 2 ({package_name}_NEW_GEMINIA):
```sql
{proc_code2}
```


Your task is to analyze these two versions and produce a detailed comparison report. Follow these steps:

1. Carefully examine both versions of the procedure.
2. Identify all differences between the two versions.
3. Categorize the differences into the following areas:
   - Changes in Logic
   - Changes in Parameters
   - Changes in SQL Statements
   - Changes in Exception Handling
   - Other Changes
4. For each identified difference, consider its potential impact on the procedure's functionality and performance.
5. Provide clear, actionable insights that would help an intern merge these two procedures effectively.

Before writing your final report, break down your comparison process in <comparison_process> tags. This will help ensure a thorough and accurate analysis. In this section:

1. List the main sections of each procedure side by side.
2. Identify and quote specific differences in each section.
3. Categorize each difference according to the provided categories.
4. Consider potential impacts of each difference.

Pay special attention to changes in business logic, as these are critical for preventing regression issues.

After your analysis, provide your final report in Markdown format with the following structure:

1. ## Introduction
   - Brief overview of the comparison task
   - Names of the procedures being compared

2. ## Summary of Differences
   - ### Changes in Logic
   - ### Changes in Parameters
   - ### Changes in SQL Statements
   - ### Changes in Exception Handling
   - ### Other Changes

3. ## Detailed Comparison
   - For each significant difference:
     - Explain what changed
     - Discuss potential implications
     - Provide guidance for merging (if applicable)

4. ## Conclusion
   - Overall assessment of the changes
   - Recommendations for merging the procedures

Ensure your report is clear, concise, and focused on helping the team understand the differences and merge the procedures effectively.
"""

    try:
        response = client.chat.completions.create(
                    model="gemini-1.5-flash",
                    messages=[
                        {"role": "system", "content": "You are an expert PL/SQL developer."},
                        {"role": "user", "content": prompt}
                    ],
                    temperature=0.3,
                    #max_tokens=2000,
                    n=1,
                    #stop=None,
        )
        print("RESPONSE FROM LLMs", response.choices[0].message.content)
        report = response.choices[0].message.content
        return report
    except Exception as e:
        logging.error(f"Error during Geminia API call for component '{proc_name}': {e}")
        return None
````

````python
def load_components_from_directory(base_dir, package_name, package_variant):
    """
    Loads PL/SQL components from the specified package variant directory.

    Parameters:
    - base_dir: Base directory where package directories are located (e.g., 'packages')
    - package_variant: Variant of the package ('HERITAGE_BODY' or 'GEMINIA_BODY')

    Returns:
    - components: A dictionary with component types as keys and dictionaries of component names and codes as values.
    """
    components = {
        'procedures': {},
        'functions': {},
        'cursors': {},
        'types': {},
        'variables': {},
    }

    # Construct the full path to the package variant directory
    package_dir = os.path.join(base_dir, f'{package_name}_{package_variant}')
    
    if not os.path.isdir(package_dir):
        logging.error(f"Directory '{package_dir}' does not exist.")
        return components

    logging.info(f"Loading components from '{package_dir}'.")

    for comp_type in components.keys():
        comp_type_dir = os.path.join(package_dir, comp_type)
        if not os.path.isdir(comp_type_dir):
            logging.warning(f"Component type directory '{comp_type_dir}' does not exist. Skipping.")
            continue

        for filename in os.listdir(comp_type_dir):
            if filename.endswith('.sql'):
                comp_name = os.path.splitext(filename)[0].upper()  # Normalize to uppercase
                file_path = os.path.join(comp_type_dir, filename)
                try:
                    with open(file_path, 'r', encoding='utf-8') as file:
                        code = file.read().strip()
                        components[comp_type][comp_name] = code
                        logging.debug(f"Loaded {comp_type[:-1].capitalize()} '{comp_name}' from '{file_path}'.")
                except Exception as e:
                    logging.error(f"Failed to read file '{file_path}': {e}")

    logging.info(f"Loaded components from '{package_variant}'.")
    return components
````

````python
def main():
    # Define base directory and package variants
    base_directory = 'packages2'
    heritage_variant = 'HERITAGE_BODY'
    geminia_variant = 'NEW_GEMINIA_BODY'
    package_name = 'GIN_STP_PKG'  # Replace with your actual package name

    # Load components from HERITAGE and GEMINIA variants
    logging.info("Starting to load HERITAGE components.")
    heritage_components = load_components_from_directory(base_directory, package_name, heritage_variant)
    
    logging.info("Starting to load GEMINIA components.")
    geminia_components = load_components_from_directory(base_directory, package_name, geminia_variant)

    # Proceed only if components are loaded
    if not any(heritage_components.values()):
        logging.error("No HERITAGE components loaded. Aborting comparison.")
        return
    if not any(geminia_components.values()):
        logging.error("No GEMINIA components loaded. Aborting comparison.")
        return

    # Compare the components using GPT-4
    logging.info("Starting comparison of components using GPT-4.")
    differences = compare_components_gpt4(heritage_components, geminia_components, package_name)

    # Output the differences
    logging.info("Comparison completed. Summary of differences:")
    for comp_type, diff in differences.items():
        logging.info(f"\nComponent Type: {comp_type.capitalize()}")
        logging.info(f"Added: {diff['added']}")
        logging.info(f"Removed: {diff['removed']}")
        logging.info(f"Modified: {diff['modified']}")

    # Optionally, you can save or process the 'differences' dictionary as needed

````
